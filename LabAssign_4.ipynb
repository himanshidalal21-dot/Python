```python
import numpy as np

A = np.random.randint(1, 10, (3, 3))
B = np.random.randint(1, 10, (3, 3))

print("Matrix A:\n", A)
print("\nMatrix B:\n", B)

add = A + B
print("\nAddition (A + B):\n", add)

mul = np.dot(A, B)
print("\nMultiplication (A x B):\n", mul)
```

```python
import numpy as np

print("Enter elements for 5x3 matrix:")
A = []
for i in range(5):
    row = list(map(int, input().split()))
    A.append(row)

print("Enter elements for 3x2 matrix:")
B = []
for i in range(3):
    row = list(map(int, input().split()))
    B.append(row)

A = np.array(A)
B = np.array(B)

# Multiply
product = np.dot(A, B)

print("\nProduct Matrix (5x2):\n")
print(product)
```
