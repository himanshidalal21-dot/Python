```python
def maximum(a,b,c):
    return max(a,b,c)

print(maximum(10,5,8))
```

```python
def sum_list(lst):
    return sum(lst)

print(sum_list([8,2,3,0,7]))
```

```python
def multiply_list(lst):
    result=1
    for i in lst:
        result*=i
    return result

print(multiply_list([8,2,3,-1,7]))
```

```python
def reverse_string(s):
    return s[::-1]

print(reverse_string("1234abcd"))
```

```python
def factorial(n):
    f=1
    for i in range(1,n+1):
        f*=i
    return f

print(factorial(5))
```

```python
def check_range(n):
    if n in range(1,11):
        print("In Range")
    else:
        print("Not in Range")

check_range(5)
```

```python
def count_case(s):
    upper=0
    lower=0
    for i in s:
        if i.isupper():
            upper+=1
        elif i.islower():
            lower+=1
    print("Upper case:",upper)
    print("Lower case:",lower)

count_case("The quick Brow Fox")
```

```python
def unique_list(lst):
    return list(set(lst))

print(unique_list([1,2,3,3,3,3,4,5]))
```

```python
def prime(n):
    for i in range(2,n):
        if n%i==0:
            return "Not Prime"
    return "Prime"

print(prime(7))
```

```python
def even_list(lst):
    return [i for i in lst if i%2==0]

print(even_list([1,2,3,4,5,6,7,8,9]))
```

```python
def perfect(n):
    s=0
    for i in range(1,n):
        if n%i==0:
            s+=i
    if s==n:
        print("Perfect Number")
    else:
        print("Not Perfect")

perfect(6)
```

```python
def palindrome(s):
    if s==s[::-1]:
        print("Palindrome")
    else:
        print("Not Palindrome")

palindrome("madam")
```

```python
def pascal(n):
    for i in range(n):
        num=1
        for j in range(i+1):
            print(num,end=" ")
            num=num*(i-j)//(j+1)
        print()

pascal(5)
```
