```python
str = 'SBJain College 54678'
space = 0
digit = 0
alpha = 0
na = 0
for i in str:
  if (i.isspace()):
    space += 1
  if (i.isdigit()):
    digit += 1
  if (i.isalpha()):
    alpha += 1
  if (i.isalnum()):
    na += 1
print("Count Space: ",space)
print("Count Digit: ",digit)
print("Count Alphabet: ",alpha)
print("Count Digit and Alphabet: ",na)
```
