```python
def string_stats(s):
    vowels = "aeiouAEIOU"
    num_vowels = sum(1 for c in s if c in vowels)
    num_consonants = sum(1 for c in s if c.isalpha() and c not in vowels)
    num_spaces = s.count(' ')
    num_lowercase = sum(1 for c in s if c.islower())

    print("Number of vowels:", num_vowels)
    print("Number of consonants:", num_consonants)
    print("Number of spaces:", num_spaces)
    print("Number of lowercase letters:", num_lowercase)

# Example usage:
text = input("Enter a string: ")
string_stats(text)
```

```python
def capitalize_lines():
    print("Enter lines (type 'END' to finish):")
    lines = []
    while True:
        line = input()
        if line == "END":
            break
        lines.append(line.upper())

    print("\nCapitalized lines:")
    for line in lines:
        print(line)

capitalize_lines()
```
